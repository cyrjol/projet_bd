```python
def dataset(i, path):
    #i : scenario's number
    #path : where the data is save 
    n_11=pd.read_csv(path+f'Scenario-{i}/Demands/Node_11.csv')
    n_11 =n_11.drop(n_11.columns[0], axis=1)
    
    l_21=pd.read_csv(path+f'Scenario-{i}/Flows/Link_21.csv')
    l_21 =l_21.drop(l_21.columns[0], axis=1)

    
    l_22=pd.read_csv(path+f'Scenario-{i}/Flows/Link_22.csv')
    l_22 =l_22.drop(l_22.columns[0], axis=1)
    
    p_10=pd.read_csv(path+f'Scenario-{i}/Pressures/Node_10.csv')
    p_10 =p_10.drop(p_10.columns[0], axis=1)
    
    #merge the features into the same dataset
    df = n_11.merge(l_21, how='left', left_index=True, right_index=True)
    df = df.merge(l_22, how='left', left_index=True, right_index=True)
    df = df.merge(p_10, how='left', left_index=True, right_index=True)
    
    #scaling
    scaler = StandardScaler()
    scaler.fit(df)
    df=pd.DataFrame(scaler.transform(df))
    
    #rename the columns
    columns=['Node 11','Link 21','Link 22','Pressure 10']
    df.columns=columns
    
    #introduce the timestamp
    time = pd.read_csv(path+f'Scenario-{i}/Timestamps.csv')
    df.insert(0,"timestamp" , pd.to_datetime(time['Timestamp']))
    
    #save the dataset and load it with the timestamp as index
    df.to_csv(path+f'Scenario-{i}/df.csv',index=False)
    df=pd.read_csv(path+f'Scenario-{i}/df.csv',parse_dates=['timestamp'], index_col='timestamp')
    
    return df
```

```python
def labels(i,path):
    #i : scenario's number
    #path : where the data is save
    labels =pd.read_csv(path+f'Scenario-{i}/Labels.csv')
    labels =labels.drop(labels.columns[0], axis=1)
    
    time = pd.read_csv(path+f'Scenario-{i}/Timestamps.csv')
    labels.insert(0,"timestamp" , pd.to_datetime(time['Timestamp']))
    
    labels.to_csv(path+f'Scenario-{i}/y.csv')
    return labels
```

```python
for i in {1,2,3,4,5,6,7,10,22,505}:
    dataset(i,path)
    labels(i,path)
```
